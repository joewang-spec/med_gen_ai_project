```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm  # Progress bar

# ==========================================
# 1. CONFIGURATION
# ==========================================
CONFIG = {
    "IMG_SIZE": (128, 128),       # Resizing images to 128x128
    "BATCH_SIZE": 16,
    "LEARNING_RATE": 1e-4,
    "EPOCHS": 15,                 # Increased slightly for better convergence
    "DEVICE": "cuda" if torch.cuda.is_available() else "cpu",
    "DATA_ROOT": "./data/stage1_train/", # Path to your unzipped Kaggle data
    "SAVE_PLOT_PATH": "ablation_results.png"
}

print(f"Running on device: {CONFIG['DEVICE']}")

# ==========================================
# 2. DATASET (DSB2018 KAGGLE LOADER)
# ==========================================
class DSB2018Dataset(Dataset):
    def __init__(self, root_dir):
        self.root_dir = root_dir
        # Get list of image IDs (folder names)
        self.image_ids = next(os.walk(root_dir))[1]
        print(f"Found {len(self.image_ids)} images in {root_dir}")

    def __len__(self):
        return len(self.image_ids)

    def __getitem__(self, idx):
        img_id = self.image_ids[idx]
        base_path = os.path.join(self.root_dir, img_id)
        
        # 1. Load Image
        img_path = os.path.join(base_path, "images", f"{img_id}.png")
        # Convert to RGB (some medical images are RGBA or Grayscale)
        image = Image.open(img_path).convert("RGB")
        
        # 2. Load Masks (Merge all individual nucleus masks into one)
        mask_dir = os.path.join(base_path, "masks")
        # Create empty mask of same size as image
        combined_mask = np.zeros((image.size[1], image.size[0]), dtype=np.uint8)
        
        for mask_file in os.listdir(mask_dir):
            if mask_file.endswith('.png'):
                m_path = os.path.join(mask_dir, mask_file)
                m = np.array(Image.open(m_path).convert("L"))
                combined_mask = np.maximum(combined_mask, m) # Logical OR
            
        mask = Image.fromarray(combined_mask)

        # 3. Apply Transforms (Resize & Tensor)
        # Note: In a full paper, you'd want data augmentation here.
        resize = transforms.Resize(CONFIG["IMG_SIZE"])
        to_tensor = transforms.ToTensor()

        image = resize(image)
        mask = resize(mask)
        
        image = to_tensor(image)
        mask = to_tensor(mask)

        return image, mask

# ==========================================
# 3. THE ABLATABLE U-NET
# ==========================================
class DoubleConv(nn.Module):
    """(Convolution => [BN] => ReLU) * 2"""
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):
        return self.double_conv(x)

class HackableUNet(nn.Module):
    """
    A U-Net where you can turn skip connections ON/OFF.
    active_skips: list of 4 booleans [Skip1, Skip2, Skip3, Skip4]
                  (Skip1 = Shallowest/High-Res, Skip4 = Deepest/Low-Res)
    """
    def __init__(self, n_channels=3, n_classes=1, active_skips=[True, True, True, True]):
        super().__init__()
        self.active_skips = active_skips

        # --- ENCODER ---
        self.inc = DoubleConv(n_channels, 64)
        self.down1 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(64, 128))
        self.down2 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(128, 256))
        self.down3 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(256, 512))
        self.down4 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(512, 1024)) # Bottleneck

        # --- DECODER ---
        self.up1 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.conv1 = DoubleConv(1024, 512) 
        
        self.up2 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.conv2 = DoubleConv(512, 256)
        
        self.up3 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.conv3 = DoubleConv(256, 128)
        
        self.up4 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.conv4 = DoubleConv(128, 64)
        
        self.outc = nn.Conv2d(64, n_classes, kernel_size=1)

    def forward(self, x):
        # Encoder
        x1 = self.inc(x)       
        x2 = self.down1[1](self.down1[0](x1)) 
        x3 = self.down2[1](self.down2[0](x2)) 
        x4 = self.down3[1](self.down3[0](x3)) 
        x5 = self.down4[1](self.down4[0](x4)) 

        # Decoder with Ablation Logic
        # Note: active_skips[0] is shallowest (x1), active_skips[3] is deepest (x4)
        
        # Block 1 (Deepest)
        x = self.up1(x5)
        # Skip form x4
        if self.active_skips[3]: x = torch.cat([x4, x], dim=1)
        else: x = torch.cat([torch.zeros_like(x4), x], dim=1) # Zero fill
        x = self.conv1(x)

        # Block 2
        x = self.up2(x)
        # Skip form x3
        if self.active_skips[2]: x = torch.cat([x3, x], dim=1)
        else: x = torch.cat([torch.zeros_like(x3), x], dim=1)
        x = self.conv2(x)

        # Block 3
        x = self.up3(x)
        # Skip form x2
        if self.active_skips[1]: x = torch.cat([x2, x], dim=1)
        else: x = torch.cat([torch.zeros_like(x2), x], dim=1)
        x = self.conv3(x)

        # Block 4 (Shallowest)
        x = self.up4(x)
        # Skip form x1
        if self.active_skips[0]: x = torch.cat([x1, x], dim=1)
        else: x = torch.cat([torch.zeros_like(x1), x], dim=1)
        x = self.conv4(x)

        return self.outc(x)

# ==========================================
# 4. UTILS: DICE SCORE
# ==========================================
def dice_coeff(pred, target):
    """Calculates Dice Coefficient for validation."""
    smooth = 1.0
    pred = torch.sigmoid(pred)
    pred = (pred > 0.5).float()
    intersection = (pred * target).sum()
    return (2. * intersection + smooth) / (pred.sum() + target.sum() + smooth)

# ==========================================
# 5. EXPERIMENT RUNNER
# ==========================================
def run_ablation_study():
    # 1. Setup Data
    if not os.path.exists(CONFIG["DATA_ROOT"]):
        raise FileNotFoundError(f"Data not found at {CONFIG['DATA_ROOT']}. Please download and unzip stage1_train.zip")

    full_dataset = DSB2018Dataset(CONFIG["DATA_ROOT"])
    
    # Split: 90% Train, 10% Val
    train_size = int(0.9 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size])
    
    train_loader = DataLoader(train_dataset, batch_size=CONFIG["BATCH_SIZE"], shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=CONFIG["BATCH_SIZE"], shuffle=False)

    # 2. Define Variants (The Core Investigation)
    experiments = {
        "Full U-Net":   [True, True, True, True],
        "No Skips":     [False, False, False, False],
        "Shallow Only": [True, True, False, False], # Keeps fine detail, loses semantic
        "Deep Only":    [False, False, True, True]  # Keeps semantic, loses fine detail
    }

    trained_models = {}
    final_scores = {}

    print("\n=== STARTING ABLATION STUDY ===")
    
    for name, skips in experiments.items():
        print(f"\n>>> Training Variant: {name} | Skips: {skips}")
        model = HackableUNet(active_skips=skips).to(CONFIG["DEVICE"])
        
        # Standard optimizer
        optimizer = optim.Adam(model.parameters(), lr=CONFIG["LEARNING_RATE"])
        # BCEWithLogitsLoss is standard for binary segmentation
        criterion = nn.BCEWithLogitsLoss()

        for epoch in range(CONFIG["EPOCHS"]):
            # Train
            model.train()
            epoch_loss = 0
            for imgs, masks in tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['EPOCHS']}", leave=False):
                imgs, masks = imgs.to(CONFIG["DEVICE"]), masks.to(CONFIG["DEVICE"])
                
                optimizer.zero_grad()
                logits = model(imgs)
                loss = criterion(logits, masks)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()
            
            # Validate
            model.eval()
            val_dice = 0
            with torch.no_grad():
                for v_imgs, v_masks in val_loader:
                    v_imgs, v_masks = v_imgs.to(CONFIG["DEVICE"]), v_masks.to(CONFIG["DEVICE"])
                    v_logits = model(v_imgs)
                    val_dice += dice_coeff(v_logits, v_masks).item()
            
            avg_val_dice = val_dice / len(val_loader)
            print(f"    [Epoch {epoch+1}] Loss: {epoch_loss/len(train_loader):.4f} | Val Dice: {avg_val_dice:.4f}")
        
        trained_models[name] = model
        final_scores[name] = avg_val_dice

    print("\n=== FINAL SCORES ===")
    for name, score in final_scores.items():
        print(f"{name}: {score:.4f}")

    return trained_models, val_loader

# ==========================================
# 6. VISUALIZATION & REPORT GENERATION
# ==========================================
def save_visual_comparison(models_dict, loader):
    print(f"\nGenerating comparison image at {CONFIG['SAVE_PLOT_PATH']}...")
    
    # Grab one batch from validation
    imgs, masks = next(iter(loader))
    # Pick the first image in the batch that actually has cells (some are empty)
    idx = 0 
    
    sample_img = imgs[idx].unsqueeze(0).to(CONFIG["DEVICE"])
    ground_truth = masks[idx].squeeze().cpu().numpy()

    # Plotting
    num_variants = len(models_dict)
    fig, axes = plt.subplots(1, num_variants + 2, figsize=(20, 5))
    
    # 1. Original Image
    axes[0].imshow(imgs[idx].permute(1, 2, 0).cpu().numpy())
    axes[0].set_title("Input Image")
    axes[0].axis('off')
    
    # 2. Ground Truth
    axes[1].imshow(ground_truth, cmap='gray')
    axes[1].set_title("Ground Truth")
    axes[1].axis('off')

    # 3. Predictions from each variant
    for i, (name, model) in enumerate(models_dict.items()):
        model.eval()
        with torch.no_grad():
            pred = torch.sigmoid(model(sample_img))
            # Binarize for visualization
            pred_mask = (pred > 0.5).float().cpu().numpy()[0, 0]
        
        ax = axes[i + 2]
        ax.imshow(pred_mask, cmap='gray')
        ax.set_title(f"{name}")
        ax.axis('off')
    
    plt.tight_layout()
    plt.savefig(CONFIG["SAVE_PLOT_PATH"])
    plt.show()
    print("Done!")

if __name__ == "__main__":
    # Run everything
    models, v_loader = run_ablation_study()
    save_visual_comparison(models, v_loader)
```

```python
# ==========================================
# 4. UPDATED UTILS: DETAILED METRICS
# ==========================================
def compute_batch_metrics(pred_logits, target_mask):
    """
    Computes Precision, Recall, Dice, and IoU for a batch of predictions.
    """
    # Apply sigmoid and threshold to get binary predictions
    preds = (torch.sigmoid(pred_logits) > 0.5).float()
    
    # Flatten tensors to simplify calculation (batch_size * channels * h * w)
    preds = preds.view(-1)
    targets = target_mask.view(-1)
    
    # Calculate True Positives, False Positives, False Negatives
    TP = (preds * targets).sum()
    FP = (preds * (1 - targets)).sum()
    FN = ((1 - preds) * targets).sum()
    
    smooth = 1e-6 # Small constant to prevent division by zero
    
    precision = TP / (TP + FP + smooth)
    recall = TP / (TP + FN + smooth)
    dice = (2 * TP) / (2 * TP + FP + FN + smooth)
    iou = TP / (TP + FP + FN + smooth)
    
    return {
        "Precision": precision.item(),
        "Recall": recall.item(),
        "Dice": dice.item(),
        "IoU": iou.item()
    }

# ==========================================
# 6. NEW SECTION: VISUALIZATION & REPORTING
# ==========================================

def visualize_dataset_samples(loader, num_samples=5):
    """
    Visualizes a few training samples (Image vs Ground Truth) to verify data loading.
    """
    imgs, masks = next(iter(loader))
    
    plt.figure(figsize=(15, 6))
    plt.suptitle("Dataset Samples (Top: Input, Bottom: Ground Truth)", fontsize=16)
    
    for i in range(min(num_samples, len(imgs))):
        # Plot Image
        plt.subplot(2, num_samples, i + 1)
        # Permute (C, H, W) -> (H, W, C) for matplotlib
        plt.imshow(imgs[i].permute(1, 2, 0).cpu().numpy()) 
        plt.axis('off')
        
        # Plot Mask
        plt.subplot(2, num_samples, i + 1 + num_samples)
        plt.imshow(masks[i].squeeze().cpu().numpy(), cmap='gray')
        plt.axis('off')
        
    plt.tight_layout()
    plt.show()

def evaluate_models_comprehensive(models_dict, val_loader):
    """
    Runs a full evaluation on the validation set and prints a comparison table.
    """
    results_table = {}
    
    print("\n" + "="*60)
    print(f"{'Model Variant':<20} | {'Dice':<8} | {'IoU':<8} | {'Precision':<10} | {'Recall':<8}")
    print("-" * 60)
    
    for name, model in models_dict.items():
        model.eval()
        metrics_sum = {"Precision": 0, "Recall": 0, "Dice": 0, "IoU": 0}
        num_batches = 0
        
        with torch.no_grad():
            for v_imgs, v_masks in val_loader:
                v_imgs, v_masks = v_imgs.to(CONFIG["DEVICE"]), v_masks.to(CONFIG["DEVICE"])
                v_logits = model(v_imgs)
                
                # Get metrics for this batch
                batch_metrics = compute_batch_metrics(v_logits, v_masks)
                
                for k, v in batch_metrics.items():
                    metrics_sum[k] += v
                num_batches += 1
        
        # Average over all batches
        avg_metrics = {k: v / num_batches for k, v in metrics_sum.items()}
        results_table[name] = avg_metrics
        
        print(f"{name:<20} | {avg_metrics['Dice']:.4f}   | {avg_metrics['IoU']:.4f}   | {avg_metrics['Precision']:.4f}     | {avg_metrics['Recall']:.4f}")
    
    print("="*60 + "\n")
    return results_table

def save_visual_comparison_grid(models_dict, loader):
    """
    Saves a visual comparison of predictions from all models.
    """
    print(f"Generating comparison image...")
    
    # Grab one batch
    imgs, masks = next(iter(loader))
    
    # Find an index with a non-empty mask for a better visual
    idx = 0
    for i in range(len(masks)):
        if masks[i].sum() > 0:
            idx = i
            break
            
    sample_img = imgs[idx].unsqueeze(0).to(CONFIG["DEVICE"])
    ground_truth = masks[idx].squeeze().cpu().numpy()

    # Create Plot
    num_models = len(models_dict)
    fig, axes = plt.subplots(1, num_models + 2, figsize=(4 * (num_models + 2), 4))
    
    # 1. Input Image
    axes[0].imshow(imgs[idx].permute(1, 2, 0).cpu().numpy())
    axes[0].set_title("Input Image", fontsize=14)
    axes[0].axis('off')
    
    # 2. Ground Truth
    axes[1].imshow(ground_truth, cmap='gray')
    axes[1].set_title("Ground Truth", fontsize=14)
    axes[1].axis('off')
    
    # 3. Model Predictions
    for i, (name, model) in enumerate(models_dict.items()):
        model.eval()
        with torch.no_grad():
            pred_logits = model(sample_img)
            # Get metrics just for this specific image for the title
            single_metrics = compute_batch_metrics(pred_logits, masks[idx].unsqueeze(0).to(CONFIG["DEVICE"]))
            
            pred_mask = (torch.sigmoid(pred_logits) > 0.5).float().cpu().numpy()[0, 0]
        
        ax = axes[i + 2]
        ax.imshow(pred_mask, cmap='gray')
        # Title with Dice score for quick reference
        ax.set_title(f"{name}\nDice: {single_metrics['Dice']:.2f}", fontsize=12)
        ax.axis('off')
    
    plt.tight_layout()
    save_path = CONFIG.get("SAVE_PLOT_PATH", "comparison.png")
    plt.savefig(save_path)
    plt.show()
    print(f"Comparison saved to {save_path}")

# ==========================================
# EXECUTION BLOCK (Update your __main__)
# ==========================================
if __name__ == "__main__":
    # 1. Load Data
    full_dataset = DSB2018Dataset(CONFIG["DATA_ROOT"])
    
    # Split Data
    train_size = int(0.9 * len(full_dataset))
    val_size = len(full_dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(full_dataset, [train_size, val_size])
    
    train_loader = DataLoader(train_dataset, batch_size=CONFIG["BATCH_SIZE"], shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=CONFIG["BATCH_SIZE"], shuffle=False)

    # 2. VISUALIZE DATASET (New Step)
    print("Visualizing random dataset samples...")
    visualize_dataset_samples(train_loader)

    # 3. Run Training (Uses your existing function, just ensure it returns 'trained_models')
    trained_models, _ = run_ablation_study()
    
    # 4. EVALUATE & VISUALIZE RESULTS (New Section)
    print("Running comprehensive evaluation...")
    evaluate_models_comprehensive(trained_models, val_loader)
    
    print("Saving visual comparisons...")
    save_visual_comparison_grid(trained_models, val_loader)
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
from tqdm.notebook import tqdm # Use notebook progress bar
import matplotlib.pyplot as plt

# ==========================================
# 1. CONFIGURATION
# ==========================================
CONFIG = {
    # RAW string for Windows path
    "DATA_ROOT": r"C:\Users\joewa\files\cmu\genai\data\Kvasir-SEG", 
    "IMG_FOLDER": "images",  
    "MASK_FOLDER": "masks",
    
    "IMG_SIZE": (256, 256),       
    "BATCH_SIZE": 8,            
    "LEARNING_RATE": 1e-4,
    "EPOCHS": 15, # Enough to show divergence between models
    "DEVICE": "cuda" if torch.cuda.is_available() else "cpu",
}

print(f"Running on: {CONFIG['DEVICE']}")

# ==========================================
# 2. DATASET LOADER (Robust Medical)
# ==========================================
class MedicalSegmentationDataset(Dataset):
    def __init__(self, root_dir, img_folder, mask_folder):
        self.root_dir = root_dir
        self.image_dir = os.path.join(root_dir, img_folder)
        self.mask_dir = os.path.join(root_dir, mask_folder)
        
        valid_exts = ('.jpg', '.png', '.jpeg')
        self.images = [f for f in os.listdir(self.image_dir) if f.lower().endswith(valid_exts)]
        
        if len(self.images) == 0:
            print(f"WARNING: No images found in {self.image_dir}")

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_filename = self.images[idx]
        
        # 1. Load Image
        img_path = os.path.join(self.image_dir, img_filename)
        image = Image.open(img_path).convert("RGB")
        
        # 2. Load Mask (Robust Search)
        # Kvasir masks usually have same name but might differ in extension
        mask_path = os.path.join(self.mask_dir, img_filename)
        if not os.path.exists(mask_path):
             mask_path = os.path.join(self.mask_dir, img_filename.replace(".jpg", ".png"))
        
        # If still not found, try to find ANY file with the same basename
        if not os.path.exists(mask_path):
             basename = os.path.splitext(img_filename)[0]
             for f in os.listdir(self.mask_dir):
                 if basename in f:
                     mask_path = os.path.join(self.mask_dir, f)
                     break

        mask = Image.open(mask_path).convert("L")

        # 3. Transform
        resize = transforms.Resize(CONFIG["IMG_SIZE"])
        to_tensor = transforms.ToTensor()

        image = to_tensor(resize(image))
        mask = to_tensor(resize(mask))
        mask = (mask > 0.5).float() # Binarize

        return image, mask

# ==========================================
# 3. METRICS CALCULATOR
# ==========================================
def calculate_metrics(pred_logits, target_mask):
    """
    Computes Precision, Recall, Dice, and IoU for a batch.
    """
    preds = (torch.sigmoid(pred_logits) > 0.5).float()
    
    # Flatten
    preds = preds.view(-1)
    targets = target_mask.view(-1)
    
    TP = (preds * targets).sum()
    FP = (preds * (1 - targets)).sum()
    FN = ((1 - preds) * targets).sum()
    
    smooth = 1e-6
    
    precision = TP / (TP + FP + smooth)
    recall = TP / (TP + FN + smooth)
    dice = (2 * TP) / (2 * TP + FP + FN + smooth)
    iou = TP / (TP + FP + FN + smooth)
    
    return {
        "Dice": dice.item(),
        "IoU": iou.item(),
        "Precision": precision.item(),
        "Recall": recall.item()
    }

# ==========================================
# 4. HACKABLE U-NET MODEL
# ==========================================
class DoubleConv(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.double_conv = nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.BatchNorm2d(out_channels),
            nn.ReLU(inplace=True)
        )
    def forward(self, x): return self.double_conv(x)

class HackableUNet(nn.Module):
    def __init__(self, active_skips=[True, True, True, True]):
        super().__init__()
        self.active_skips = active_skips # [Shallowest, ..., Deepest]

        self.inc = DoubleConv(3, 64)
        self.down1 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(64, 128))
        self.down2 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(128, 256))
        self.down3 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(256, 512))
        self.down4 = nn.Sequential(nn.MaxPool2d(2), DoubleConv(512, 1024))

        self.up1 = nn.ConvTranspose2d(1024, 512, kernel_size=2, stride=2)
        self.conv1 = DoubleConv(1024, 512)
        self.up2 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.conv2 = DoubleConv(512, 256)
        self.up3 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.conv3 = DoubleConv(256, 128)
        self.up4 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.conv4 = DoubleConv(128, 64)
        self.outc = nn.Conv2d(64, 1, kernel_size=1)

    def forward(self, x):
        x1 = self.inc(x)
        x2 = self.down1[1](self.down1[0](x1))
        x3 = self.down2[1](self.down2[0](x2))
        x4 = self.down3[1](self.down3[0](x3))
        x5 = self.down4[1](self.down4[0](x4))

        # Decoder with Switches
        x = self.up1(x5)
        if self.active_skips[3]: x = torch.cat([x4, x], dim=1)
        else: x = torch.cat([torch.zeros_like(x4), x], dim=1)
        x = self.conv1(x)

        x = self.up2(x)
        if self.active_skips[2]: x = torch.cat([x3, x], dim=1)
        else: x = torch.cat([torch.zeros_like(x3), x], dim=1)
        x = self.conv2(x)

        x = self.up3(x)
        if self.active_skips[1]: x = torch.cat([x2, x], dim=1)
        else: x = torch.cat([torch.zeros_like(x2), x], dim=1)
        x = self.conv3(x)

        x = self.up4(x)
        if self.active_skips[0]: x = torch.cat([x1, x], dim=1)
        else: x = torch.cat([torch.zeros_like(x1), x], dim=1)
        x = self.conv4(x)

        return self.outc(x)

# ==========================================
# 5. MAIN EXPERIMENT LOOP
# ==========================================
def run_experiments():
    # 1. Load Data
    dataset = MedicalSegmentationDataset(CONFIG["DATA_ROOT"], CONFIG["IMG_FOLDER"], CONFIG["MASK_FOLDER"])
    print(f"Dataset Size: {len(dataset)} images")
    
    # Split
    train_size = int(0.85 * len(dataset))
    val_size = len(dataset) - train_size
    train_ds, val_ds = torch.utils.data.random_split(dataset, [train_size, val_size])
    
    train_loader = DataLoader(train_ds, batch_size=CONFIG["BATCH_SIZE"], shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=CONFIG["BATCH_SIZE"], shuffle=False)

    # 2. Define The 4 Variants
    variants = {
        "No Skips":       [False, False, False, False],
        "Deep Only":      [False, False, True, True],
        "Shallow Only":   [True, True, False, False],
        "Full U-Net":     [True, True, True, True]
    }
    
    results_stats = []
    trained_models = {}

    # 3. Train Each Variant
    for name, skips in variants.items():
        print(f"\nTraining Variant: {name}...")
        model = HackableUNet(active_skips=skips).to(CONFIG["DEVICE"])
        optimizer = optim.Adam(model.parameters(), lr=CONFIG["LEARNING_RATE"])
        criterion = nn.BCEWithLogitsLoss()
        
        # Training
        model.train()
        for epoch in range(CONFIG["EPOCHS"]):
            epoch_loss = 0
            # Use tqdm for progress bar
            for img, mask in tqdm(train_loader, desc=f"{name} Epoch {epoch+1}", leave=False):
                img, mask = img.to(CONFIG["DEVICE"]), mask.to(CONFIG["DEVICE"])
                optimizer.zero_grad()
                out = model(img)
                loss = criterion(out, mask)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()
        
        # Evaluation
        print(f"Evaluating {name}...")
        model.eval()
        metrics_sum = {"Dice": 0, "IoU": 0, "Precision": 0, "Recall": 0}
        num_batches = 0
        
        with torch.no_grad():
            for img, mask in val_loader:
                img, mask = img.to(CONFIG["DEVICE"]), mask.to(CONFIG["DEVICE"])
                out = model(img)
                batch_metrics = calculate_metrics(out, mask)
                
                for k, v in batch_metrics.items():
                    metrics_sum[k] += v
                num_batches += 1
        
        # Average Stats
        avg_stats = {k: v/num_batches for k, v in metrics_sum.items()}
        avg_stats["Variant"] = name
        results_stats.append(avg_stats)
        trained_models[name] = model

    return pd.DataFrame(results_stats), trained_models, val_loader

# ==========================================
# 6. VISUALIZATION FUNCTION
# ==========================================
def plot_comparisons(models, loader):
    img, mask = next(iter(loader))
    # Find a non-empty mask for better visualization
    idx = 0
    for i in range(len(mask)):
        if mask[i].sum() > 0: idx = i; break
    
    sample = img[idx].unsqueeze(0).to(CONFIG["DEVICE"])
    
    plt.figure(figsize=(15, 4))
    
    # 1. Input
    plt.subplot(1, 6, 1)
    plt.imshow(img[idx].permute(1,2,0).cpu())
    plt.title("Input")
    plt.axis('off')
    
    # 2. GT
    plt.subplot(1, 6, 2)
    plt.imshow(mask[idx].squeeze().cpu(), cmap='gray')
    plt.title("Ground Truth")
    plt.axis('off')
    
    # 3. Models
    model_order = ["No Skips", "Deep Only", "Shallow Only", "Full U-Net"]
    for i, name in enumerate(model_order):
        if name in models:
            models[name].eval()
            with torch.no_grad():
                pred = torch.sigmoid(models[name](sample))
                pred = (pred > 0.5).float().cpu().numpy()[0,0]
            
            plt.subplot(1, 6, i+3)
            plt.imshow(pred, cmap='gray')
            plt.title(name)
            plt.axis('off')
            
    plt.tight_layout()
    plt.show()

# ==========================================
# RUN EVERYTHING
# ==========================================
if __name__ == "__main__":
    # Run Experiments
    df_results, models, v_loader = run_experiments()
    
    # Display Stats Table
    print("\n=== FINAL PAPER STATS ===")
    # Reorder columns
    df_results = df_results[["Variant", "Dice", "IoU", "Precision", "Recall"]]
    display(df_results) # Pretty print for Jupyter
    
    # Display Visuals
    print("\n=== VISUAL COMPARISON ===")
    plot_comparisons(models, v_loader)
```
